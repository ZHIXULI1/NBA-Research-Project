```python
import csv
with open("superman_Q1_BBALL.csv",'w') as f:
    writer = csv.writer(f)
    writer.writerow(["Game_ID","Player_ID","Player_Plus/Minus"])
every_game = play_by_play.groupby("Game_id")

for game_id, df in every_game:
    #     print df
    # df =play_by_play[play_by_play["Game_id"]== "021fd159b55773fba8157e2090fe0fe2"]#need to be substituted by groupby method
    sorted_table = df.sort_values(by=["Period","PC_Time", "WC_Time","Event_Num"], ascending=[True,False,True,True])
    game_line_up = lineup[lineup["Game_id"] == game_id]
    period_1 = game_line_up[game_line_up["Period"] == 1]
    all_players = {} #output all players that have been on the court
    team_players_dict = {} #current players on the court
    temp_dict = {}
    # team1 = []
    # team2 = []
    period_count = 1
    count = 1
    for index, row in period_1.iterrows():
    #     team1.append(row[""])
        if row["Team_id"] not in team_players_dict:
            team_players_dict[row["Team_id"]] = []
        team_players_dict[row["Team_id"]].append(row["Person_id"])
        all_players[row["Person_id"]] = 0

    for index, row in sorted_table.iterrows():
        if row["Period"] == period_count:
            if row["Event_Msg_Type"] == 1:
    #             team_id = row["Team_id"]
                shooter = row["Person1"]
                team_id = ""
                for key, value in team_players_dict.iteritems():
                        if shooter in value:
                            team_id = key
    #             print row["Event_Num"]
    #           team_players_dict
                #check which persons are on the court for each team.
                #first, for each game and each team, track who are on the court (use the raw data in lineup for each period,
                #to create to players list, and track for event_msg_type 8 substitution. Create a large set to track all the
                #players in the game.)
                #for each team plus or minus corresponding score differntial
                score = row["Option1"]
    #             print row["Event_Num"]
    #             print all_players
                for key,value in team_players_dict.iteritems():
                    for item in value:
    #                     print all_players[item]
                        if key == team_id:
                            all_players[item] +=score
                        else:
                            all_players[item] -=score
            elif row["Event_Msg_Type"] == 3:
                if row["Option1"] == 1:
                    #get point
                    shooter = row["Person1"]
                    for key, value in temp_dict.iteritems():
                            if shooter in value:
                                team_id = key
                    for key,value in temp_dict.iteritems():
                        for item in value:
                            if key == team_id:
                                all_players[item] +=1
                            else:
                                all_players[item] -=1
            elif row["Event_Msg_Type"] == 8: # update current list & all_player 
                #important: for event_Msg_type 6 we need to check the time, if the time is large than foul but less than 
                #a free throw, we should keep that person for the free throw.
                #for that substitution, we don't consider the second person.

                shooter = row["Person1"]
                for key, value in team_players_dict.iteritems():
                        if shooter in value:
                            team_id = key
    #             print row["Event_Num"]
                team_players_dict[team_id].append(row["Person2"])

                team_players_dict[team_id].remove(row["Person1"])
    #             print team_players_dict[team_id]
                if row["Person2"] not in all_players:
                    all_players[row["Person2"]] = 0
            elif row["Event_Msg_Type"] == 6:
                #after a foul, only 3 or 8 is made if there is a free throw. We can check that. if there is a 3 or 8,
                #get a temporary list of players
                i =index
    #             print i
                for r in range(1, 11):
    #                 print r
                    if i+r >= len(sorted_table)-1:
                        break
                    elif sorted_table.loc[i+r, 'Period'] != period_count:
                        break
                    elif sorted_table.loc[i+r,'Event_Msg_Type'] == 6:
                        break
                    elif sorted_table.loc[i+r,'Event_Msg_Type'] == 3:
    #                         free_throw = true
                        temp_dict = team_players_dict
        else:
            period_count += 1
    #         print period_count
            next_period = game_line_up[game_line_up["Period"] == period_count]
    #         print next_period
            team_players_dict.clear()
    #         print team_players_dict
            for index, row in next_period.iterrows():
                if row["Team_id"] not in team_players_dict:
                    team_players_dict[row["Team_id"]] = []
                team_players_dict[row["Team_id"]].append(row["Person_id"])
                if row["Person_id"] not in all_players:
                    all_players[row["Person_id"]] = 0
            #update player dict
    with open("teeeeam_Q1_BBALL.csv",'a') as f2:
        writer = csv.writer(f2)
        keys = all_players.keys()
        values = all_players.values()
        for num in range(len(all_players)):
            writer.writerow([game_id,keys[num],values[num]])        
        
    
```
